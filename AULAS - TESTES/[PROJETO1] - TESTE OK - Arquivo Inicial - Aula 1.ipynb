### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a direto, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
# vamos precisar instalar:
# pyautogui -> automações de mouse e teclado
# pandas -> manipulação e análise de dados
# openpyxl -> trabalhar com arquivos excel (xlsx)

# para instalar: abrir o terminal e rodar:
# pip install pyautogui pandas openpyxl

# pyautogui click -> clicar com o mouse o mouse
# pyautogui.write -> escrever com o teclado
# pyautogui.press -> apertar uma tecla
# pyautogui.hotkey -> apertar uma combinação de teclas
```

```python
import pyautogui
import time

pyautogui.PAUSE = 1  # pausa de 1 segundo entre as ações

# Passo a passo do desafio
# Passo 1: Entrar no sistema da empresa (link do drive) # abrir o Chrome
# abri o chrome
pyautogui.hotkey("win","r")
pyautogui.write("chrome https://google.com.br")
pyautogui.press("enter")

time.sleep(1)  # esperar o chrome abrir

# focar barra de endereço
pyautogui.hotkey("ctrl", "l")
time.sleep(0.2)

# digitei o link do sistema
link = "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga"
pyautogui.write(link)

# apertei enter e esperei carregar
pyautogui.press("enter")
time.sleep(3)   
```

```python
# Passo 2: Navegar no sistema para encontrar a base de dados
time.sleep(1)
pyautogui.click(x=490, y=542, clicks=2)  # clica no arquivo *****
time.sleep(1)

# Passo 3: Exportar a base de dados (baixar o arquivo)
pyautogui.click(x=506, y=532) #clica no arquivo
pyautogui.click(x=838, y=419) # clica em baixar/exportar
time.sleep(0.5)  # esperar o download terminar

# PODERIA SER UM CAMINHO FIXO TAMBÉM, MAS VAMOS PARA O PROXIMO PASSO NA PASTA DE DOWNLOADS
# pyautogui.click(x=-162, y=267) # clicar na barra de downloads
# pyautogui.click(x=-370, y=383) # clicar no arquivo baixado
# fechar o chrome
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
import pandas

# Passo 4: Calcular os indicares (faturamento e quantidade de produtos vendidos)
# abrir a base de dados
caminho = r"C:\Users\gaabi\Downloads\Vendas - Dez.xlsx"
tabela = pandas.read_excel(caminho)


# ver as informações da base de dados
display(tabela) # mostra a tabela no jupyter notebook ou pode usar print(não é tão visual)


# somar o faturamento de todos os produtos
faturamento = tabela["Valor Final"].sum()

# somar a quantidade de produtos vendidos
qtde_produtos = tabela["Quantidade"].sum()

print(faturamento)
print(qtde_produtos)

```

### Vamos agora enviar um e-mail pelo gmail

```python
import pyperclip
import time

# Passo 5: Enviar um e-mail para a diretoria com os indicadores

# abrir o ggmail no chrome opção 1
# pyautogui.hotkey("win","r")
# pyautogui.write("chrome https://google.com.br")
# pyautogui.press("enter")
# time.sleep(2)  # esperar o chrome abrir

# focar barra de endereço
# pyautogui.hotkey("ctrl", "l")
# time.sleep(0.2)

# digitei o link do gmail
# link = "https://mail.google.com/mail/u/0/"
# pyautogui.write(link)
# pyautogui.press("enter")

# abrir o gmail no chrome opção 2
from pyautogui import click


pyautogui.hotkey("ctrl", "t")  # abrir nova aba
time.sleep(2)  # esperar o chrome abrir

link = "https://mail.google.com/"
pyautogui.write(link)
pyautogui.press("backspace")  # limpar o possível texto automático
pyautogui.press("enter")

pyautogui.click(x=165, y=254)  # clicar em escrever 
time.sleep(3.5)  # esperar carregar a tela de escrever e-mail

# escrever o e-mail
# para quem
# destinatário
pyautogui.write("gaabi-campos+codigo@hotmail.com")
pyautogui.press("enter")   # FIXA o destinatário (muito importante)
time.sleep(0.3)

pyautogui.press("tab")     # vai pro assunto
time.sleep(0.2)

# qual o assunto
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl", "v")
time.sleep(0.2)

pyautogui.press("tab")     # vai pro corpo
time.sleep(0.2)

# qual o corpo do e-mail
texto = f"""
Prezados,

Segue o relatório de vendas de hoje.
Faturamento: R${faturamento:,.2f}
Quantidade de produtos vendidos: {qtde_produtos:,}

Qualquer dúvida estou à disposição.
Atenciosamente,
Gabriella Campos
"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")
time.sleep(0.2)

# enviar o e-mail
pyautogui.click(x=1218, y=977)  # clicar em enviar o e-mail
```
